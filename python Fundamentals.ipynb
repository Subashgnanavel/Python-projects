```python
total_sum = 0
while True:
    number = int(input("Enter an integer (negative number to stop): "))
    if number < 0:
        print("Negative number entered. Stopping input.")
        break
    total_sum += number
print("Total sum of all entered numbers:", total_sum)
```

```python
def check_even_odd(number):
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"
print(check_even_odd(4))
print(check_even_odd(7))

```

```python
def reverse_string(s):
    return s[::-1]
print(reverse_string("hello"))
print(reverse_string("Subash"))
```

```python
def max_of_three(x, y, z):
    return max(x, y, z)
print(max_of_three(6, 12, 9))
print(max_of_three(2, 3, 7))
```

```python
def count_vowels(s):
    vowels = "aeiouAEIOU"
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count
print(count_vowels("hello"))
print(count_vowels("Subash Gnanavel"))
```
