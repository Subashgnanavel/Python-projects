```python

name = "Hello Kgisl"
name = name.lower()
vowels = "aeiou"
count = 0
for char in name:
    if char in vowels:
        count += 1
print("Number of vowels:", count)
```

```python
squares = [0] * 10
for i in range(1, 11):
    squares[i - 1] = i * i
print("List of squares:", squares)
```

```python
words = ['dog','tiger','lion','cat']
long_words = [word for word in words if len(word) > 3]
print("Words with more than 3 letters:", long_words)
```
